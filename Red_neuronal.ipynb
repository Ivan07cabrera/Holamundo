```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import RMSprop, SGD
from tensorflow.keras import regularizers
import numpy as np
```

```python
#Parametros de la red neuronal
learning_rate = 0.001
epochs = 20
batch_size = 128
```

```python
#Datos de mnist
dataset=mnist.load_data()
(x_train, y_train), (x_test, y_test) = dataset
```

```python
x_train.shape  #Mostrar el tamaño de los datos de entrenamiento
```

```python
x_trainv = x_train.reshape(60000, 784)
x_testv = x_test.reshape(10000, 784)  #Aplanar imagenes
```

```python
x_trainv = x_trainv.astype('float32')
x_testv = x_testv.astype('float32')  #Covertir datos a flotantes
```

```python
x_trainv /= 255.   #Normalizar datos
x_testv /= 255.
```

```python
num_classes = 10
y_trainc = keras.utils.to_categorical(y_train, num_classes)   #one hot encoding
y_testc = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = Sequential()
model.add(Dense(512, activation='sigmoid', input_shape=(784,)))
model.add(Dense(num_classes, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss=keras.losses.CategoricalCrossentropy(from_logits=True),optimizer=SGD(learning_rate=learning_rate),metrics=['accuracy'])
```

```python
modelo = tf.keras.models.load_model("red.keras")
```

```python
history = model.fit(x_trainv, y_trainc,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_testv, y_testc)
                    )
```

```python
model.save("red.keras")
print("Modelo guardado correctamente")
```
